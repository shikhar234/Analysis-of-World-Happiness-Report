## Multicolinearity test by VIF

```python
inVar=['Economy_GDP_per_Capita','Family','Health(Life_Expectancy)','Freedom','Trust(Government_Corruption)','Dystopia_Residual']

from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) 
    for i in range(X.shape[1])]
    return(vif)
```

```python
X =df[['Economy_GDP_per_Capita','Family','Health(Life_Expectancy)','Freedom','Trust(Government_Corruption)','Dystopia_Residual']]
calc_vif(X)
```

we can see that VIF values for Economy_GDP_per_Capita,Family,Health(Life_Expectancy),Freedom is greater then 10 so above variables are highly correlated and multicolinearity exist in between them

removing Health(Life_Expectancy)

```python
X =df[['Economy_GDP_per_Capita','Freedom','Family','Trust(Government_Corruption)','Dystopia_Residual']]
calc_vif(X)
```

```python
import statsmodels.api as sm
inVar=['Economy_GDP_per_Capita','Freedom','Trust(Government_Corruption)','Dystopia_Residual']
Z =df[inVar]
Y = df['Happiness_Score']
Z = sm.add_constant(Z)
result = sm.OLS(Y, Z).fit()
print(result.summary())
```

WE can see that the R square value is 0.844 and adjusted R square value is also 0.843 
where the p values for all consider variables is <0.01
